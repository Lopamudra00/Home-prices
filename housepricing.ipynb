```python
import pandas as pd
```

```python
df = pd.read_csv('houseprice.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('Price(US $)')
plt.scatter(df.area, df.price)

```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']],df.price)

```

```python
reg.predict([[3300]])
```
